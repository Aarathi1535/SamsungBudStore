```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://en.wikipedia.org/wiki/Samsung_Galaxy#Devices"
```

```python
resp = requests.get(url)
```

```python
print(resp.content)
```

```python
soup = BeautifulSoup(resp.content,'html.parser')
print(soup)
```

```python
print(soup.prettify())
```

```python
table = soup.find("table",{"class":"wikitable"})
```

```python
mobile_names = []
```

```python
for row in table.find_all("tr")[1:]:
    columns = row.find_all("td")
    if columns:
        name = columns[0].get_text(strip=True)
        mobile_names.append(name)
```

```python
print(mobile_names)
```

```python
mobile_names[0] = mobile_names[0][:19]
```

```python
mobile_names[1] = mobile_names[1][:20]
mobile_names[2] = mobile_names[2][:24]
mobile_names[3] = mobile_names[3][:23]

```

```python
mobile_names
```

```python
dict_of_mobiles = {}
```

```python
dict_of_mobiles[mobile_names[0]] = 17999
dict_of_mobiles[mobile_names[1]] = 8990
dict_of_mobiles[mobile_names[2]] = 4990
dict_of_mobiles[mobile_names[3]] = 9990
dict_of_mobiles[mobile_names[4]] = 11990
dict_of_mobiles[mobile_names[5]] = 17990
```

```python
dict_of_mobiles
```

```python
import csv
cost = []
for i in dict_of_mobiles.keys():
        cost.append(dict_of_mobiles[i])
combined = zip(mobile_names,cost)
with open(r"C:\Users\hp\OneDrive\Desktop\example.csv", "w", newline="") as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(["Mobile Name","Cost"])
    for data in combined:
        writer.writerow(data)
```
